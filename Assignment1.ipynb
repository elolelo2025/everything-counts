```python
# Visualisation 1

# Convert Date to datetime
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Time Series Plot: Average Price Over Time
plt.figure()
df_sorted = df.sort_values("Date")
plt.plot(df_sorted["Date"], df_sorted["AveragePrice"])
plt.title("Average Price Over Time")
plt.xlabel("Date")
plt.ylabel("AveragePrice")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisation 2

# Histogram of Average Price
plt.figure()
plt.hist(df["AveragePrice"].dropna().values)
plt.title("Distribution of Average Price")
plt.xlabel("AveragePrice")
plt.ylabel("Frequency")
plt.show()


```

```python
# Visualisation 3

# Scatter Plot: FTBPrice vs FOOPrice (First-time buyers vs Former owners)
if "FTBPrice" in df.columns and "FOOPrice" in df.columns:
    plt.figure()
    plt.scatter(df["FTBPrice"].dropna().values[:1000], df["FOOPrice"].dropna().values[:1000])
    plt.title("FTBPrice vs FOOPrice")
    plt.xlabel("FTBPrice")
    plt.ylabel("FOOPrice")
    plt.show()
```

```python
# Visualisation 4

# Confidence Interval Plot of average price over time (yearly)

# Yearly mean + CI
df["Year"] = df["Date"].dt.year
yearly = df.groupby("Year")["AveragePrice"].agg(["mean", "count", "std"])
yearly["sem"] = yearly["std"] / np.sqrt(yearly["count"])
yearly["ci95"] = 1.96 * yearly["sem"]

plt.figure(figsize=(8, 5))
plt.plot(yearly.index, yearly["mean"])
plt.fill_between(
    yearly.index,
    yearly["mean"] - yearly["ci95"],
    yearly["mean"] + yearly["ci95"],
    alpha=0.3
)
plt.title("Average Price Yearly Mean with 95% Confidence Interval")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.tight_layout()
plt.show()
```

```python
# Visualisation 5

# Histogram of average price
plt.figure(figsize=(8, 5))
plt.hist(df["AveragePrice"].dropna(), bins=50)
plt.title("Histogram of Average Price")
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Visualisation 6

# Box plot of prices across property types
price_cols = ["DetachedPrice", "SemiDetachedPrice", "TerracedPrice", "FlatPrice"]
subset = df[price_cols].dropna()

plt.figure(figsize=(8, 5))
plt.boxplot([subset[col] for col in price_cols], labels=price_cols)
plt.title("Distribution of Prices by Property Type")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```
